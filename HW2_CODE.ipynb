```python
import numpy as np
import matplotlib.pyplot as plt
import math

x=np.linspace(-3,3,50)
y=np.cos((3*x**2-5*x)**3)
yPrim=-np.sin((3*x**2-5*x)**3)*(3*(3*x**2-5*x)**2)*(6*x-5)

plt.plot(x, y, label='Original', linewidth=2.0)
plt.plot(x, yPrim,label='First Derivative', linewidth=2.0)
plt.legend()
plt.show()
```
